```python
install.packages("bnlearn")
library(bnlearn)

# Assuming all these variables are problematic and need conversion
variables_to_convert <- c("EC100", "EC160", "IT101", "IT161", "MA101", "PH100", "PH160", "HS101", "QP")

# Convert character variables to factors
course.grades[variables_to_convert] <- lapply(course.grades[variables_to_convert], as.factor)

course.grades.net<-hc(course.grades[,-9],score="k2")
plot(course.grades.net)
```

```python
course.grades.net.fit<-bn.fit(course.grades.net, course.grades[,-9])

course.grades.net.fit$EC100
course.grades.net.fit$EC160
course.grades.net.fit$IT101
course.grades.net.fit$IT161
course.grades.net.fit$PH100
course.grades.net.fit$PH160
course.grades.net.fit$MA101
course.grades.net.fit$HS101

bn.fit.barchart(course.grades.net.fit$EC100)
bn.fit.barchart(course.grades.net.fit$EC160)
bn.fit.barchart(course.grades.net.fit$IT101)
bn.fit.barchart(course.grades.net.fit$IT161)
bn.fit.barchart(course.grades.net.fit$PH100)
bn.fit.barchart(course.grades.net.fit$PH160)
bn.fit.barchart(course.grades.net.fit$MA101)
bn.fit.barchart(course.grades.net.fit$HS101)
```

```python
course.grades.PH100 <- data.frame(cpdist(course.grades.net.fit,
                                         nodes = "PH100",
                                         evidence = ((EC100=="DD")&(IT101=="CC")&(MA101=="CD"))))

library(dplyr)
df <- course.grades.PH100 %>%
  group_by(PH100) %>%
  summarise(counts = n())

library(ggplot2)
ggplot(df, aes(x = PH100, y = counts)) +geom_bar(fill = "#0073C2FF", stat = "identity") +geom_text(aes(label = counts), vjust = -0.3)
```

```python
library(e1071)
library(visNetwork)
library(bnlearn)

# Read the dataset
course.grades <- read.table("2020_bn_nb_data.txt", header = TRUE)

# Assuming all these variables are problematic and need conversion
variables_to_convert <- c("EC100", "EC160", "IT101", "IT161", "MA101", "PH100", "PH160", "HS101", "QP")

# Convert character variables to factors
course.grades[variables_to_convert] <- lapply(course.grades[variables_to_convert], as.factor)

# Initialize vector to store accuracy values
accuracy_values <- numeric(20)

# Loop to run the analysis 20 times
for (i in 1:20) {
  set.seed(i) # Different seed for each iteration
  # Split data into training and testing sets
  train_indices <- sample(1:nrow(course.grades), size = round(0.7 * nrow(course.grades)))
  train_data <- course.grades[train_indices, ]
  test_data <- course.grades[-train_indices, ]

  nb <- naive.bayes(train_data, "QP")

  # Plot the NB model
  graphviz.plot(nb)

  fitted_nb <- bn.fit(nb, train_data, method = "bayes")

  pred <- predict(fitted_nb, test_data)

  cm <- table(predicted = pred, true = test_data$QP)

  cat("Confusion Matrix for iteration", i, ":\n")
  print(cm)

  accuracy_values[i] <- sum(diag(cm)) / sum(cm)

  cat("Accuracy for iteration", i, " is : ",accuracy_values[i], "\n\n")
}

# Print minimum, maximum, and average accuracy
cat("Minimum Accuracy:", min(accuracy_values), "\n")
cat("Maximum Accuracy:", max(accuracy_values), "\n")
cat("Average Accuracy:", mean(accuracy_values), "\n")
```

```python
library(e1071)
library(visNetwork)
library(bnlearn)

# Read the dataset
course.grades <- read.table("2020_bn_nb_data.txt", header = TRUE)

# Assuming all these variables are problematic and need conversion
variables_to_convert <- c("EC100", "EC160", "IT101", "IT161", "MA101", "PH100", "PH160", "HS101", "QP")

# Convert character variables to factors
course.grades[variables_to_convert] <- lapply(course.grades[variables_to_convert], as.factor)

# Initialize vector to store accuracy values
accuracy_values <- numeric(20)

# Loop to run the analysis 20 times
for (i in 1:20) {
  set.seed(i*i*i*i) # Different seed for each iteration
  # Split data into training and testing sets
  train_indices <- sample(1:nrow(course.grades), size = round(0.7 * nrow(course.grades)))
  train_data <- course.grades[train_indices, ]
  test_data <- course.grades[-train_indices, ]

  # Build TAN model
  tan <- tree.bayes(train_data, "QP")

  # Plot the TAN model
  graphviz.plot(tan)

  # Fit TAN model
  fitted_tan <- bn.fit(tan, train_data, method = "bayes")

  # Make predictions on testing data
  pred <- predict(fitted_tan, test_data)

  # Generate contingency table
  cm <- table(predicted = pred, true = test_data$QP)
  print(cm)

  accuracy_values[i] <- sum(diag(cm)) / sum(cm)
  cat("Accuracy for iteration", i, " is : ",accuracy_values[i], "\n\n")
}

# Print minimum, maximum, and average accuracy

cat("Minimum Accuracy:", min(accuracy_values), "\n")
cat("Maximum Accuracy:", max(accuracy_values), "\n")
cat("Average Accuracy:", mean(accuracy_values), "\n")
```
